# clean data

```python
import pandas as pd
```

```python
input_path = 'data/census.csv'
output_path = 'data/clean_census.csv'
```

```python
# read data by removing whitespaces
df = pd.read_csv(input_path, skipinitialspace=True)
df.head()
```

```python
df_cleaned = df.replace("?", None)
df_imputed = df_cleaned.dropna()
```

```python
df_imputed.head()
```

```python
# save cleaned and imputed data
df_imputed.to_csv(output_path, index=False)
```
